# RGB LED med formodstande og MicroPython p en ESP32

### Introduktion
Denne tutorial vil vise, hvordan du tilslutter og styrer en RGB LED med formodstande ved hjlp af MicroPython p en ESP32. Du vil lre at installere ndvendige biblioteker, opstte forbindelser og skrive kode til at styre LED'ens farver.



### Materialer
- ESP32
- RGB LED
- Formodstande (f.eks. 220)
- Ledninger
- Breadboard
- Thonny IDE

**1. Forbindelser**

Forbind din RGB LED til din ESP32 som flger:

- Forbind den lange ben (katode) p RGB LED'en til GND.
- Forbind de rde, grnne og bl ben til GPIO21, GPIO22, og GPIO23 gennem formodstande.
- Forbindelsesskema
 
```markdown
RGB LED   ESP32
------------------
R         GPIO21
G         GPIO22
B         GPIO23
GND       GND
```

**2. Skriv dit frste script**
```python
from machine import Pin, PWM
from time import sleep

# Opstning af pins til RGB LED
red = PWM(Pin(21), freq=1000)
green = PWM(Pin(22), freq=1000)
blue = PWM(Pin(23), freq=1000)

# Funktion til at stte farven p LED'en
def set_color(r, g, b):
    red.duty(int(r * 1023 / 255))
    green.duty(int(g * 1023 / 255))
    blue.duty(int(b * 1023 / 255))

# Farveskift loop
while True:
    set_color(255, 0, 0)  # Rd
    sleep(1)
    set_color(0, 255, 0)  # Grn
    sleep(1)
    set_color(0, 0, 255)  # Bl
    sleep(1)
    set_color(255, 255, 0)  # Gul
    sleep(1)
    set_color(0, 255, 255)  # Cyan
    sleep(1)
    set_color(255, 0, 255)  # Magenta
    sleep(1)
    set_color(255, 255, 255)  # Hvid
    sleep(1)
    set_color(0, 0, 0)  # Slukket
    sleep(1)
```

**3. Upload og kr scriptet**

- Klik p den grnne "Run" knap i Thonny eller tryk F5.
- Gem scriptet p ESP32 som main.py.

**4. Forstelse af koden**
- `from machine import Pin, PWM`: Importerer `Pin` og `PWM` klasserne fra `machine` modulet.
- `from time import sleep`: Importerer `sleep` funktionen fra time modulet.
- `red = PWM(Pin(21), freq=1000)`: Initialiserer PWM for rd kanal p pin GPIO21 med en frekvens p 1000 Hz.
- `green = PWM(Pin(22), freq=1000)`: Initialiserer PWM for grn kanal p pin GPIO22.
- `blue = PWM(Pin(23), freq=1000)`: Initialiserer PWM for bl kanal p pin GPIO23.
- `set_color(r, g, b)`: Funktion til at stte farven p LED'en ved at justere PWM duty cycle.
- `while True`: Starter en uendelig lkke for at skifte farver.
