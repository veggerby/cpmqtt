#### **3. Strenge og basisoperationer:**

- **Strengeoperationer:**

"Strenge er tekst, og vi kan gre mange ting med dem, ligesom at sammenkde dem, gentage dem, og tage enkelte tegn ud. Her er nogle eksempler:"

---
```Python
tekst1 = "Hej"
tekst2 = " Verden"
fuld_tekst = tekst1 + tekst2    # Sammenkdning
gentaget_tekst = tekst1 * 3     # Gentagelse
frste_bogstav = tekst1[0]      # Indeksering
```

"Lister er som en samling af variabler, som vi kan gemme i n variabel. Vi kan tilfje, fjerne og f adgang til elementer i en liste. Her er nogle eksempler:"

---
```Python
liste = [1, 2, 3, 4]
liste.append(5)        # Tilfje element
liste.remove(2)        # Fjerne element
frste_element = liste[0]  # Adgang til element
```
