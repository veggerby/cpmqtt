### Opgave Hjælp
#### 1. **Variabler og Datatyper:**

- **Variabler:**

"Lad os starte med variabler. En variabel er som en kasse, hvor vi kan gemme information. Vi giver kassen et navn, så vi kan huske, hvad vi har gemt der. For eksempel:"

---
```Python
navn = "Alice"
alder = 10
```
---

"Her har vi to variabler, 'navn' og 'alder'. 'Navn' gemmer en tekst, og 'alder' gemmer et tal."

- **Datatyper:**

"Der findes forskellige typer data, som vi kan gemme i variabler. De vigtigste er:

Heltal (int), som er hele tal.
Flydende tal (float), som er tal med decimaler.
Strenge (str), som er tekst.
Lister (list), som kan indeholde flere værdier."


- **Kontrol af variabler:**

"For at se, hvad der er gemt i variablerne, kan man få dem udskrevet ved at bruge kommandoen print. Det er her, kassens navn bliver anvendeligt."

"Her er nogle eksempler:"

---
```Python
heltal = 5
print(heltal)
flydende_tal = 3.14
print(flydende_tal)
tekst = "Hej Pirater - Her er en meget laaaaaaang tekst man ikke vil skrive flere gange, da man også risikere stave fejl. Derfor er variabler mega smarte"
print(tekst)
liste = [1, 2, 3, 4]
print(liste)
```

#### **3. Strenge og basisoperationer:**

- **Strengeoperationer:**

"Strenge er tekst, og vi kan gøre mange ting med dem, ligesom at sammenkæde dem, gentage dem, og tage enkelte tegn ud. Her er nogle eksempler:"

---
```Python
tekst1 = "Hej"
tekst2 = " Verden"
fuld_tekst = tekst1 + tekst2    # Sammenkædning
gentaget_tekst = tekst1 * 3     # Gentagelse
første_bogstav = tekst1[0]      # Indeksering
```
---

"Vi kan bruge '+' til at sammenkæde to strenge, '*' til at gentage en streng flere gange, og '[]' til at få et bestemt tegn fra strengen."

- **Lister og basisoperationer:**
  
"Lister er som en samling af variabler, som vi kan gemme i én variabel. Vi kan tilføje, fjerne og få adgang til elementer i en liste. Her er nogle eksempler:"

---
```Python
liste = [1, 2, 3, 4]
liste.append(5)        # Tilføje element
liste.remove(2)        # Fjerne element
første_element = liste[0]  # Adgang til element
```
---

"Med '.append()' kan vi tilføje et element til listen, med '.remove()' kan vi fjerne et element, og med '[]' kan vi få adgang til et bestemt element i listen."
