#### **4. Funktioner:**

"Nu skal vi tale om funktioner. En funktion er et stykke kode, som vi kan genbruge flere gange i vores program. Det hj√¶lper os med at organisere vores kode og g√∏re den mere overskuelig. Vi definerer en funktion √©n gang og kan derefter kalde den, n√•r vi har brug for den."

- **Definition og Kald af Funktioner:**
"Lad os starte med at se, hvordan vi definerer en funktion. Vi bruger n√∏gleordet 'def' efterfulgt af funktionens navn og parenteser. Inden i parenteserne kan vi skrive parametre, som er information, vi kan give til funktionen. Herefter kommer et kolon og indrykket kode, som er selve funktionens krop."

---
```Python
def sig_hej():
    print("Hej Verden!")
```
---

"Her har vi defineret en simpel funktion, som hedder 'sig_hej'. N√•r vi kalder funktionen, udskriver den 'Hej Verden!'"

- **Kalder Funktioner:**
  
"Lad os nu se, hvordan vi kalder en funktion. Det er meget simpelt - vi skriver bare funktionens navn efterfulgt af parenteser."

---
```Python
sig_hej()  # Dette kalder funktionen og udskriver "Hej Verden!"
```
---

- **Funktioner med Parametre:**

"Vi kan ogs√• lave funktioner, der tager imod parametre. Det betyder, at vi kan give funktionen forskellige v√¶rdier, som den kan arbejde med."

---
```Python
def sig_hej_til(navn):
    print(f"Hej {navn}!")
```
---

"Her har vi defineret en funktion, der hedder 'sig_hej_til', som tager √©t parameter, 'navn'. N√•r vi kalder funktionen, udskriver den 'Hej' efterfulgt af det navn, vi giver den."

---
```Python
sig_hej_til("Alice")  # Udskriver "Hej Alice!"
sig_hej_til("Bob")    # Udskriver "Hej Bob!"
```
---

- **Returnere V√¶rdier fra Funktioner:**

"Funktioner kan ogs√• returnere v√¶rdier ved hj√¶lp af n√∏gleordet 'return'. Det betyder, at funktionen kan give os et resultat, som vi kan bruge videre i vores program."

#### **5. Loops:**

"Nu skal vi tale om loops. Et loop er et stykke kode, gentager sig selv flere gange i vores program. Loops kan hj√¶lpe os til ikke at skulle skive den samme kode flere gange, hvis vi har en sekvens der skal gentages."

- **For-Loop:**

"Et for-loop kan feks. k√∏re et fast antal gange, eller bruges til at gentages for hvert element der er i en liste:

```python
# For-Loop der k√∏re 6 gange
for x in range(6):
  print(x)
```

```python
# For-Loop der k√∏re en gang for hvert element i listen
liste = [1, 2, 3, 4]
for tal in liste:
  print(tal)
```

- **While-Loop**

"Et While-Loop k√∏re s√• l√¶nge et statement er sandt"

```python
# While-Loop der k√∏re 5 gange
i = 1
while i < 6: # S√• l√¶nge `i` er mindre end 6
  print(i)
  i += 1
```

```python
# Uendeligt Loop
i = 1
while True: # S√• l√¶nge `True` er sandt, hmmm
  print(i)
  i += 1
```
